```python
with open('c1.txt','w') as t:
    t.write('CREATE OR REPLACE VIEW dune_user_generated.contract_address')
    t.write('\n')
    t.write('(product, contract_address) AS VALUES')
    t.write('\n')
    for i in range(len(data)):
        t.write("('")
        t.write(data.product_name[i])
        t.write("', '")
        t.write(data.contract_address[i].replace("0","\\",1))
        t.write("'::bytea")
        t.write("),")
        t.write('\n')
    
```
