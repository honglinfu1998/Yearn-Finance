```python
import json
import pandas as pd
import re
```

```python
pro_id = range(1,3)
for i in pro_id:
    with open(str(i) +'.json','r') as f:
        df = json.load(f)
        buffer = pd.DataFrame(df['data']['votes'])
        #Seperate vp_by_strategy
        vp_YFI = [str(buffer.vp_by_strategy[j][0]) for j in range(len(buffer))]
        vp_yYFI = [str(0) for j in range(len(buffer))]
        vp_DLG = [str(buffer.vp_by_strategy[j][1]) for j in range(len(buffer))]
        #Add ID column
        string_buffer = [re.sub("'", "",buffer.proposal[0]['id']) for j in range(len(buffer))]
        space_buffer = [re.sub("'", "",buffer.space[0]['id']) for j in range(len(buffer))]
        result = buffer.assign(YP_BY_YFI=vp_YFI, VP_BY_YYFI = vp_yYFI, VP_BY_DLG=vp_DLG,proposal_id=string_buffer, space = space_buffer )
        #result = buffer.assign(VP_BY_DLG=vp_DLG)
        data = pd.DataFrame().assign( proposal_id = result['proposal_id'],
                                      vote_id=result['id'], 
                                      voter_address=result['voter'], 
                                      vp = result['vp'], 
                                      created = result['created'], 
                                      choice = result['choice'], 
                                      space = result['space'], 
                                      YP_BY_YFI = result['YP_BY_YFI'], 
                                      VP_BY_YYFI = result['VP_BY_YYFI'],
                                      VP_BY_DLG = result['VP_BY_DLG'])
    #change to list for write
    pi = list(map(str, data.proposal_id))
    vi = list(map(str, data.vote_id))
    va = list(map(str, data.voter_address))
    vp = list(map(str, data.vp))
    ct = list(map(str,data.created))
    ch = list(map(str,data.choice))
    sp = list(map(str,data.space))
    yy = list(map(str,data.YP_BY_YFI)) ######typo doesn't mater
    yyy = list(map(str,data.VP_BY_YYFI))
    yd = list(map(str,data.VP_BY_DLG))
    
    #write to txt file
    with open(str(i) + '.txt','w') as t:
        t.write('CREATE OR REPLACE VIEW dune_user_generated.YVT_' + str(i))
        t.write('\n')
        t.write('(proposal_id, vote_id, voter_address, vp, created, choice, space, VP_BY_YFI, VP_BY_YYFI, VP_BY_DLG) AS VALUES')
        t.write('\n')
        for j in range(len(data)):
            t.write("('")
            t.write(pi[j])
            t.write("', '")
            t.write(vi[j])
            t.write("', '")
            t.write(va[j].replace("0","\\",1))
            t.write("'::bytea, '")
            t.write(vp[j])
            t.write("', '")
            t.write(ct[j])
            t.write("', '")
            t.write(ch[j])
            t.write("', '")
            t.write(sp[j])
            t.write("', '")
            t.write(yy[j])
            t.write("', '")
            t.write(yyy[j])
            t.write("', '")
            t.write(yd[j])
            t.write("'),")
            t.write('\n')
        
```
